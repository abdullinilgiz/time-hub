```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales = pd.read_csv('../data/shop_sales.csv')
sales.head()
```

```python
prices = pd.read_csv('../data/shop_sales_prices.csv')
prices.head()
```

```python
calendar = pd.read_csv('../data/shop_sales_dates.csv')
calendar.drop(axis=1, columns=['event_name_1', 'event_type_1',
                               'event_name_2', 'event_type_2',], inplace=True,)
calendar.head()
```

```python
data = pd.merge(pd.merge(calendar, sales, on='date_id'),
                prices, 
                on=['wm_yr_wk', 'store_id', 'item_id'], 
                how='left')
data['sell_price'] = data.groupby(
    ['store_id', 'item_id']
    )['sell_price'].transform(lambda x: x.fillna(x.mean()))
data.head()
```

```python
data.to_csv('../data/processed_data.csv')
```
